```python
# importing lib's
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing csv file into dataframe
data=pd.read_csv('world_happiness_day.csv')
data
```

158 rows and 12 columns

```python
# removing columns which are not necessary for target
data=data.drop(columns=['Country','Region','Happiness Rank','Standard Error','Dystopia Residual'])
data
```

after removing unnecessary columns we left with 7 columns which are contributing to happiness score

```python
data.round(decimals=3)
```

we get to know that in gdp,family,life,freedom,trust,generosity columns have 0 min value

```python
data['Economy (GDP per Capita)']=data['Economy (GDP per Capita)'].replace(0,data['Economy (GDP per Capita)'].mean())

data['Family']=data['Family'].replace(0,data['Family'].mean())

data['Health (Life Expectancy)']=data['Health (Life Expectancy)'].replace(0,data['Health (Life Expectancy)'].mean())

data['Freedom']=data['Freedom'].replace(0,data['Freedom'].mean())

data['Trust (Government Corruption)']=data['Trust (Government Corruption)'].replace(0,data['Trust (Government Corruption)'].mean())

data['Generosity']=data['Generosity'].replace(0,data['Generosity'].mean())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
maxacc=0
maxrs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=lr.score(x_train,y_train)
    r2 = r2_score(y_test, pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('maxacc:',maxacc,'\nmaxrs:',maxrs,'\nr squared',r2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=158)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
acc=lr.score(x_train,y_train)
r2 = r2_score(y_test, pred)
print('Accuracy:',acc)
print('\nPredicted value:\n',pred,'\n\nactual value:\n',y_test)
print('\nmean square error:',mean_squared_error(pred,y_test))
print('\nR squared score',r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
model=[DecisionTreeRegressor(),SVR(),KNeighborsRegressor(),RandomForestRegressor(),AdaBoostRegressor()]
for m in model:
    m.fit(x_train,y_train)
    acc=m.score(x_train,y_train)
    predm=m.predict(x_test)
    r2 = r2_score(y_test, predm)
    print('----------',m,'-----------')
    print('Accuracy:',acc)
    print('mean square error:',mean_squared_error(predm,y_test))
    print(r2)
    print('\n')
```

```python
 from sklearn.linear_model import Lasso,Ridge,LassoCV,RidgeCV
```

```python
lassocv=LassoCV(alphas=None,max_iter=1000,normalize=True)
lassocv.fit(x_train,y_train)
```

```python
alpha=lassocv.alpha_
alpha
```

```python
lasso_reg=Lasso(alpha)
lasso_reg.fit(x_train,y_train)
acc_=lasso_reg.score(x_train,y_train)
pred=lasso_reg.predict(x_test)
print('Accuracy:',acc_)
print('mean square error:',mean_squared_error(pred,y_test))
```

```python
ridcv=RidgeCV(alphas=np.arange(0.001,0.1,0.01),normalize=True)
ridcv.fit(x_train,y_train)
alpha=ridcv.alpha_
alpha
```

```python
rid_reg=Ridge(alpha)
rid_reg.fit(x_train,y_train)
acc__=rid_reg.score(x_train,y_train)
pred=rid_reg.predict(x_test)
print('Accuracy:',acc_)
print('mean square error:',mean_squared_error(pred,y_test))
```
